```python
import csv
import json
import requests
from bs4 import BeautifulSoup

url = "https://www.imdb.com/chart/top"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

try:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    top__250 = soup.find("tbody", attrs={"class": "lister-list"}).find_all("tr")
    film_sira = 1

    with open('imdb_top_250.csv', mode='w', newline='', encoding='utf-8') as csv_file:
        fieldnames = ['film_sira', 'adi', 'yili', 'puan']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        writer.writeheader()

        for film in top__250:
            adi = film.find("td", attrs={"class": "titleColumn"}).a.text
            yili = film.find("td", attrs={"class": "titleColumn"}).span.text.strip("()")
            puan = film.find("td", attrs={"class": "ratingColumn imdbRating"}).strong.text

            print("film sira ", film_sira)
            print("film adi : ", adi)
            print("film yili : ", yili)
            print("film puan : ", puan)

            writer.writerow({'film_sira': film_sira, 'adi': adi, 'yili': yili, 'puan': puan})

            print("\n\n")
            film_sira += 1

    with open('imdb_top_250.json', mode='w', encoding='utf-8') as json_file:
        data = []

        for film in top__250:
            adi = film.find("td", attrs={"class": "titleColumn"}).a.text
            yili = film.find("td", attrs={"class": "titleColumn"}).span.text.strip("()")
            puan = film.find("td", attrs={"class": "ratingColumn imdbRating"}).strong.text

            data.append({'film_sira': film_sira, 'adi': adi, 'yili': yili, 'puan': puan})
            film_sira += 1

        json.dump(data, json_file, ensure_ascii=False, indent=4)

except requests.exceptions.RequestException as e:
    print(e)
```

```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

url = "https://www.imdb.com/chart/top"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

try:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    top_250 = soup.find("tbody", attrs={"class": "lister-list"}).find_all("tr")
    data = []

    for film in top_250:
        adi = film.find("td", attrs={"class": "titleColumn"}).a.text
        yili = film.find("td", attrs={"class": "titleColumn"}).span.text.strip("()")
        puan = film.find("td", attrs={"class": "ratingColumn imdbRating"}).strong.text

        data.append([adi, yili, puan])

    headers = ["Film Adı", "Yıl", "Puan"]
    print(tabulate(data, headers=headers))

except requests.exceptions.RequestException as e:
    print(e)
```

```python
import requests
from bs4 import BeautifulSoup
import statistics

url = "https://www.imdb.com/chart/top"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

try:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    top_250 = soup.find("tbody", attrs={"class": "lister-list"}).find_all("tr")
    ratings = []

    for film in top_250:
        rating = float(film.find("td", attrs={"class": "ratingColumn imdbRating"}).strong.text)
        ratings.append(rating)

    mode = statistics.mode(ratings)
    median = statistics.median(ratings)
    mean = statistics.mean(ratings)

    print("Mode: ", mode)
    print("Median: ", median)
    print("Mean: ", mean)

except requests.exceptions.RequestException as e:
    print(e)
```
